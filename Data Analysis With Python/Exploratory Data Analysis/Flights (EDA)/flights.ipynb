```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('flights.csv')
```

```python
df.describe()
```

```python
counts = df['month'].value_counts()
plt.bar(counts.index, counts.values)
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Distribution of Flights by Month')
plt.show()
```

```python
grouped = df.groupby('month')['passengers'].mean()
plt.plot(grouped.index, grouped.values)
plt.xlabel('Month')
plt.ylabel('Average Number of Passengers')
plt.title('Average Number of Passengers by Month')
plt.show()
```

```python
df.isna().sum()
df.duplicated().sum()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='year', y='passengers', data=df)
```

```python
monthly_passengers = df.pivot_table(values='passengers', index='month', columns='year')

plt.figure(figsize=(14, 8))
sns.heatmap(
    monthly_passengers,
    annot=True,
    fmt='.0f',
    cmap='YlOrRd',
    linewidths=0.5,
    linecolor='white',
    cbar_kws={'label': 'Number of Passengers', 'shrink': 0.8},
    annot_kws={'size': 9}
)
plt.title('Monthly Passengers by Year', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Year', fontsize=12, labelpad=10)
plt.ylabel('Month', fontsize=12, labelpad=10)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(x='month', y='passengers', hue='year', data=df, marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.title('Monthly Passengers by Year')
plt.legend(title='Year')
plt.show()
```

```python
sns.violinplot(x='year', y='passengers', data=df)
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passengers by Year')
plt.show()
```

```python
sns.histplot(df['passengers'], bins=20, kde=True)
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.title('Distribution of Passengers')
plt.show()
```
